# get data from player totals page

```python
url='http://www.basketball-reference.com/leagues/NBA_2017_totals.html'
r = requests.get(url)
soup = BeautifulSoup(r.text, 'lxml')
```

```python
table = soup.find(id='totals_stats')
```

# pace adjustment using team pace from B-ref

```python
data['aPER'] = data.apply(lambda row: 0 if row['Tm']=='TOT' else (
    row['uPER'] * (
        np.mean(dtm['Pace'])/sum(dtm[dtm['short_name'] == row['Tm']]['Pace']))), axis=1)
```
